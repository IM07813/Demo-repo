```python
import cv2
import os
import zipfile

# Extract videos
with zipfile.ZipFile('videos.zip', 'r') as zip_ref:
    zip_ref.extractall()

# Change your current directory to extracted 'videos' directory
os.chdir('videos')

# Get all video file names
videos = [video for video in os.listdir() if video.endswith(".mp4")]

# Return to root directory
os.chdir('..')

# Create a 'frames' folder if not exist
if not os.path.exists("frames"):
    os.makedirs("frames")

# Function to save frames from video
def save_frames(video_name):
    # Open the video file
    video = cv2.VideoCapture(os.path.join('videos', video_name))

    # Index to make frame names unique
    index = 0

    # Loop through video frames
    while video.isOpened():
        # Read frame from video
        ret, frame = video.read()

        # If frame is read correctly ret is True
        if not ret:
            break

        # Every second, save frame
        if index % round(video.get(cv2.CAP_PROP_FPS)) == 0: # Assume video is at 30fps
            # Resize frame
            frame = cv2.resize(frame, (640, 360)) #resize the frame to 640x360

            # Save frame in corresponding folder, with unique name
            frame_name = os.path.join('frames', video_name.strip(".mp4"), "frame_%d.jpg" % index)
            cv2.imwrite(frame_name, frame)

        index += 1
        if cv2.waitKey(1) == ord('q'):
            break

    # Release the video object
    video.release()

# Loop through all videos
for video in videos:
    # Create a corresponding folder in 'frames' for every video
    if not os.path.exists(os.path.join('frames', video.strip(".mp4"))):
        os.makedirs(os.path.join('frames', video.strip(".mp4")))
  
    # Call the save_frames function
    save_frames(video)
```
